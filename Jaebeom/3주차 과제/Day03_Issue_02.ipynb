 # 1. 리스트 슬라이싱

 ### 나의 정리
 
 데이터프레임에서 필요한 열만을 추출할 때, 열 이름을 직접 입력해야 하는가에 대한 의문에서 생긴 고민이다.
  
 문제되는 상황은 다음과 같다
 
 ```
 a = [1,2,3,4,5]
 a[3:4],a[1]
 ```
 
 이 입력의 결과는 ([4],2) 이다.
 
 컴파일러는 단일 열 지정과 범위 지정에 있어서 다른 형식의 결과를 보여주었다.
 
 우리는 여기서 slicing 했을 때의 결과는 리스트로, 인덱싱 했을때의 결과는 정수형으로 나온다는 것을 알아낼 수 있다.
 
 ```
 column_extract = [2] + list(range(4,11))
 airbnb_remake = airbnb_db.iloc[:, column_extract]
 ```
 이런식으로 작성한다면 고민하던 문제를 해결할 수 있을 것이다.
 
 
 # 2. Pull request 버튼이 보이지 않음
 
 ### 나의 정리
 
 우리가 뭔가를 add, commit, push 하려면 파일에 변화가 생겨야만 한다. 이미 한 번 pull request 까지 작성했다가 closed 된 파일에 대해서도 역시 변화가 생기지 않는다면 다시 push 한다고 해서 pull request가 열리지는 않는다. 너무도 당연한 것이 아무것도 add하고 commit 한 것이 없기 때문에 push 했을 때 Everything up-to-date 라고 떠야만 한다.
 
 ### 해결 방안
 
 이를 해결하려면 원초적으로는 파일에 작은 변화만 주면 된다. 파일에 들어가서 코드에 영향을 주지 않는 미묘한 변화만 살짝 준다면 다시 add할 수 있고 순서대로 진행하면 될 것이다. (맞나?)
 
 
 # 3. import 구문을 하단에 쓰면 안 되는 이유
 
 ### 나의 정리
 
 좋은 프로그래머는 기계가 이해하기 쉬운 코드가 아니라 사람이 이해하기 쉬운 코드를 짠다. 어차피 기계가 이해하기 쉬운 코드라는 표현은 이상하다. 어차피 기계는 이해가 안 되면 바로 에러 메시지를 내뱉기 때문이다. 그래서 사람이 이해하기 쉬운 코드를 짜는게 중요한데 import 구문을 하단에 써버리면 그에 관련된 코드들을 먼저 접하게 된 다른 사용자들이 잠시나마 혼동할 수 있는 여지를 줄 수 있기 때문이다.
 
 
 # 4. for 구문에 i를 쓰면 안 되는 이유
 
 ### 나의 정리
 
  이도 아까 3번과 비슷한 맥락인데 i는 무언가 상징적인 의미를 담지 않은 그냥 아무개 변수이다. 때문에 i 보다는 val (value) 와 같은 의미를 대충이라도 유추할 수 있는 변수를 사용한다면 훨씬 효과적인 코드를 작성할 수 있을 것이다.
  
  
 # 5. 최적의 변수명짓기
 
 ### 나의 정리
 
  변수명을 짓는 방법에는 사람마다 차이가 있겠지만, 필요하다면 줄임말을 적절하게 사용하고, 줄바꿈 또한 적재적소에 사용하는 것이 필요할 것이다. 코드의 가독성을 위한 노력이다 보니 여러 사람과 의견을 나누는 자세가 필요해 보인다.
